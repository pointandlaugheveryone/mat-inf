## Terminal Input
Reading the terminal input in Python is a straight forward process. We simpy call function input() and geather the input in terminal.

Feel free to write anything into the terminal to see the result!

```python
terminal_input = input()
print(terminal_input)
```

Now it is important to note that any input through terminal is perceived by python as a string as can be seen in the example bellow.

Try writing numbers and see for yourself.

```python
terminal_input = input()
if isinstance(terminal_input, str):
    print("input is a string")
else:
    print("input is not a string")
```

This means that we need to learn how to convert strings to data for us more suitable, thankfully python has few functions, some of them we already know.

Let us start simple, if we know the input will be only numbers we can use the following Python function.

```python
number = '120'
number_as_int = int(number)
if isinstance(number_as_int, int):
    print(f"input is now a string")
```

However this won't work with non numbers in the string.

```python
number_with_letters = '120ab'
number_as_int = int(number_with_letters)
```

To process input you can choose to use your knowledge of loops and ifs and evaluating the specific characters of the string.

```python
some_input = "ab456dsvs98s"

digits = []
letters = []
for si in some_input:
    if si.isdigit():
        digits.append(si)
    if si.isalpha():
        letters.append(si)

print("digits: ", digits)
print("letters: ", letters)

```

You can try to use the ASCII table as well, the actual numerical representation of a character in computer, although it is not recommended to use "magic numbers" ie random numbers without explanation in the code and relying on ASCII can be a bit problematic as some systems might use different values.





Try for yourself some of the letters and values!

```python
ascii_value = ord("a")
print(ascii_value)
ascii_letter = chr(97)
print(ascii_letter)
```

Don't forget, string has a many useful functions for managing its data so dont forget to use it as well!

```python
word = "hello world"
print(word.find("world")) # will find a substring in the string and returns its index
print(word.replace("world", "class")) # will replace given substring with another
print(word.split(" ")) # will split the string using the given string to do so
print(" ".join(["hey", "how", "are", "you?"])) # will join the strings with the given string
```

Finally don't forget, you can specify what kind of input you want, both in writting and both by denying it!

```python
print("write down numbers:")
nums = input()
for num in nums:
    if num.isalpha():
        print("incorrect input!")
        break
```

## Reading File Input

Terminal input is important but reading files is as well, lets dive into the topic!

In files folder we have some files that can be read, down here we have an example on how to read a file line by line.

```python
file_name = "files/example_file"
with open(file_name) as f:
    lines = f.readlines()

print(lines)
```

Note that if reading the file line by line the new line character $\n$ remains

You can simply do pure read where the program will read the whole file into one string

```python
file_name = "files/example_file"
with open(file_name) as f:
    file = f.read()

print(file)
```

Now reading isn't the only thing you can do, you can create new files and write into them as well! By adding $"w"$ we are specifying we want to write into the file

```python
file_name = "files/new_file"
with open(file_name, "w") as f:
    f.write("hello world!")
```

Take a look into the files folder and see for yourself or just run the following command!

```python
file_name = "files/new_file"
with open(file_name) as f:
    file = f.read()

print(file)
```
